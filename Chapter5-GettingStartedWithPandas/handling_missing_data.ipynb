```python
data = DataFrame([[1., 6.5, 3.],
                 [1., NA, NA],
                 [NA, NA, NA],
                 [NA, 6.5, 3.]])
data.dropna()
```

```python
data.dropna(how='all')
```

```python
df = DataFrame(np.random.randn(7, 3))
df
```

```python
df.iloc[:4, 1] = NA
df.iloc[:2, 2] = NA
df
```

```python
df.dropna(thresh=2)
```

```python
df.fillna(0)
```

```python
df.fillna({1: 0.5,
          2: -1})
```

```python
df
```

```python
df.fillna(0, inplace=True)
df
```

```python
df = DataFrame(np.random.randn(6, 3))
df
```

```python
df.fillna(method='ffill')
```
