```python
frame = DataFrame(np.arange(12.0).reshape(4,3),
                  columns=list('bde'),
                  index=['Utah', 'Ohio', 'Texas', 'Oregan'])
frame
```

```python
series = frame.iloc[0]
series
```

```python
frame - series
```

```python
series2 = Series(range(3),
                 index=list('bef'))
series2
```

```python
frame + series2
```

```python
series3 = frame['d']
series3
```

```python
frame.sub(series3, axis=0)
```

```python
frame - series3
```
